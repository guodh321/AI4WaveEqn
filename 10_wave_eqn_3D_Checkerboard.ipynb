```python
def plot_3d_tensor(tensor, alpha = 0.15, title='3D Tensor'):
    # Create a 3D plot
    fig = plt.figure(figsize=(15, 10))

    x_size, y_size, z_size = tensor.size()

    # Create a grid of coordinates
    x, y, z = np.meshgrid(np.arange(x_size), np.arange(y_size), np.arange(z_size))

    # Plot at t = 0
    ax = fig.add_subplot(221, projection='3d')
    sc = ax.scatter(x, y, z, c=tensor.cpu().detach().numpy().flatten(), cmap='viridis', alpha=alpha)
    ax.set_title(title)
    fig.colorbar(sc, ax=ax, shrink=0.5, aspect=5)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Plot the 2D profile along the x-axis at the center of the grid
    ax = fig.add_subplot(222)
    im = ax.imshow(tensor.cpu().detach().numpy()[x_size//2, :, :].T, cmap='viridis')
    ax.set_title('2D Profile along X-axis')
    fig.colorbar(im, ax=ax)
    ax.invert_yaxis()

    # Plot the 2D profile along the y-axis at the center of the grid
    ax = fig.add_subplot(223)
    im = ax.imshow(tensor.cpu().detach().numpy()[:, y_size//2, :].T, cmap='viridis')
    ax.set_title('2D Profile along Y-axis')
    fig.colorbar(im, ax=ax)
    ax.invert_yaxis()

    # Plot the 2D profile along the z-axis at the center of the grid
    ax = fig.add_subplot(224)
    im = ax.imshow(tensor.cpu().detach().numpy()[:, :, z_size//2].T, cmap='viridis')
    ax.set_title('2D Profile along Z-axis')
    fig.colorbar(im, ax=ax)
    ax.invert_yaxis()

    plt.tight_layout()
    plt.show()
```

#### one time step - skipping layers

```python
class AI4WaveNeuralNetwork_One_Skipping_3D(nn.Module):
    def __init__(self, x_size, y_size, z_size, filter_weight):
        super(AI4WaveNeuralNetwork_One_Skipping_3D, self).__init__()

        # Specify the size of the input (batch_size, channels, depth, height, width)
        in_channels = 1

        # Specify the size of the filter/kernel
        kernel_size = filter_weight.shape[2]
        padding = (kernel_size - 1) // 2

        # Create a Conv3d layer with the specified weight, input size, and padding
        self.conv_layer = nn.Conv3d(in_channels, out_channels=1, kernel_size=kernel_size, padding=padding, bias=False)
        self.conv_layer.weight.data = filter_weight

        self.x_size = x_size
        self.y_size = y_size
        self.z_size = z_size

    def forward(self, status_tn, status_tn_1, c):
        output_list = [status_tn]

        # Calculate status_tn1 based on the provided expression
        c_status_tn = c * status_tn
        conv_term = self.conv_layer(c_status_tn.view(1, self.x_size, self.y_size, self.z_size)) + c * self.conv_layer(status_tn) - status_tn * self.conv_layer(c.view(1, self.x_size, self.y_size, self.z_size))
        
        status_tn1 = (2 * status_tn - status_tn_1) - (conv_term) * (0.1 * 0.1) / 2
        
        output_list.append(status_tn1)

        output_stacked = torch.cat(output_list, dim=1)

        return output_stacked
```

```python
def create_c_3dtensor(c_value):
    c = torch.zeros((x_size, y_size, z_size), device=device)
    for i in range(c_value.shape[0]):
        ones = torch.zeros((x_size, y_size, z_size), device=device)
        ones[:, :, i*int(x_size/c_value.shape[0]):(i+1)*int(x_size/c_value.shape[0])] = 1
        c += ones * c_value[i]
    return c
```

```python
number_rows = 5
number_cols = 5
number_depth = 5
number_all = number_rows * number_cols * number_depth
torch.manual_seed(0)
c_values_tensor = torch.randint(low=1, high=10, size=(number_all,), device=device, dtype=torch.float32, requires_grad=True)

print(c_values_tensor)

def create_checkerboard_3d(c_values_tensor, xsize, ysize, zsize, number_rows, number_cols, number_depth):
    c = torch.zeros((xsize, ysize, zsize), device=device)  # Initialize c with zeros

    for k in range(number_depth):
        for i in range(number_rows):
            for j in range(number_cols):
                    ones = torch.zeros((xsize, ysize, zsize), device=device)
                    ones[i*int(xsize/number_rows):(i+1)*int(xsize/number_rows), 
                        j*int(ysize/number_cols):(j+1)*int(ysize/number_cols),
                        k*int(zsize/number_depth):(k+1)*int(zsize/number_depth)] = c_values_tensor[k*number_rows*number_cols + i*number_rows + j]
                    c += ones

    return c

c = create_checkerboard_3d(c_values_tensor, x_size, y_size, z_size, number_rows, number_cols, number_depth)

plot_3d_tensor(c, title='C Tensor')
```
